```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = ["HackSTL 2019", "HackTheLib 2020"]
y_signups = [83, 205]
y_submissions = [16, 61]
y_judges = [5, 6]
y_ratio = [16/83, 61/205]

```

```python
def plot_metric(metric, title, ylabel, filename):
    plt.style.use('ggplot')
    plt.bar(x, metric)
    plt.title(title)
    plt.ylabel(ylabel)
    plt.xlabel("Event")
    plt.savefig(filename, bbox_inches="tight")
```

```python
plot_metric(y_signups, "Number of sign-ups by Event", "Sign-ups", "signups.png")
```

```python
plot_metric(y_submissions, "Number of submissions by Event", "Submissions", "submissions.png")
```

```python
plot_metric(y_judges, "Number of judges by Event", "Judges", "judges.png")
```

```python
plot_metric(y_ratio, "Submission-to-participants ratio by Event", "Submission-to-participants ratio", "ratio.png")
```
